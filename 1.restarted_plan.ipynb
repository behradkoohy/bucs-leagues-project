```python
import pandas as pd
import re
from itertools import combinations
import pgeocode
```

```python
# read the CSV file into a pandas DataFrame
df = pd.read_csv("bucs_results.csv")

postcodes = pd.read_csv("institute_postcodes.csv")

# print the first few rows of the DataFrame to verify that it was loaded correctly
df.head()
df = df.drop(columns=['No Score Type', 'Walkover Reason','Knockout Fixtures', 'Number of Walkovers', 'Voids', 'Blowout Game(s)', 'Zero Score?', 'Blowout to Zero', 'Team Conceding Walkover', 'Fixtures',
       'League Fixtures', 'Time'])
```

```python
postcodes = postcodes[['Modified Home', 'Venue Postcode']]
postcodes
```

```python
postcodes.columns
```

```python
# Function to remove numbers and certain words from a string
def clean_team_name(s):
    # Remove all numbers from string
    s = re.sub(r'\d+', '', s)
    # Remove "Women's" and "Men's" from string
    s = s.replace("Women's ", "").replace("Men's ", "")
    # Remove extra spaces
    s = " ".join(s.split())
    return s

def extract_integers(input_string):
    # Use a regular expression to match all integer values in the input string
    integer_matches = re.findall(r'\d+', input_string)

    # Convert the matched strings to integers and retur
    return integer_matches
```

```python
df.columns

```

```python
df22 = df[df["Community"]=="BUCS Hockey 22-23"]
```

```python
df22l = df22[df22["Fixture type"]=="League"]
```

```python
df22l
```

```python
df22l['Home Institute'] = df22l.apply(lambda row: clean_team_name(row.Home), axis=1)
df22l['Home Team'] = df22l.apply(lambda row: extract_integers(row.Home)[0], axis=1)
df22l['Away Institute'] = df22l.apply(lambda row: clean_team_name(row.Away), axis=1)
df22l['Away Team'] = df22l.apply(lambda row: extract_integers(row.Away)[0], axis=1)
```

```python
df22l
```

```python
df22l = pd.merge(left=df22l, right=postcodes, left_on="Home Institute", right_on="Modified Home")
```

```python
df22l
```

```python
latlong = pgeocode.Nominatim('GB')


def get_lat_long(postcode):
    query = latlong.query_postal_code(postcode)
    return query['latitude'], query['longitude']

get_lat_long('ox13 5gp')
```

```python
df22l['HomeLat'] = df22l.apply(lambda row: get_lat_long(row['Venue Postcode'])[0], axis=1)
df22l['HomeLon'] = df22l.apply(lambda row: get_lat_long(row['Venue Postcode'])[1], axis=1)
```

```python
dist= pgeocode.GeoDistance('GB')
```

```python
total_distances = {}

for comp in sorted(df22l.Competition.unique()):
#     print("generating combinations", comp)
#     if "Premier" in comp:
#         continue
    comp_pcd = {}
    
    competition_fixtures = df22l[df22l.Competition==comp]
    competition_fixtures = competition_fixtures[['Home', 'Venue Postcode']]

    for i, row in competition_fixtures.iterrows():
        comp_pcd[str(row[0])] = str(row[1])
    
    total_dist = 0.0
    
    combs = (list(combinations(list(comp_pcd.values()),2)))
    for comb in combs:
        d = dist.query_postal_code(comb[0], comb[1])
        total_dist += d
    
    total_dist = total_dist/len(comp_pcd)
    print(comp, "\t\t Competitors:", len(comp_pcd), "\t KMs Travelled:", total_dist)
    total_distances[comp] = total_dist
total_distances_pd = {'League': list(total_distances.keys()), "KM": list(total_distances.values())}
```

```python
avg_dists = pd.DataFrame(total_distances_pd)
avg_dists.to_csv('uncut_dists.csv', index=False)
```
